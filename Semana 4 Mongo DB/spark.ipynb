```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

# --- CONFIGURACIÓN DE SPARK PARA CARGAR DATOS DESDE MONGODB ---
spark = SparkSession.builder \
    .appName("Analisis_Categorias") \
    .config("spark.mongodb.read.connection.uri", "mongodb://database:27017/proyecto_bigdata.datos_libros") \
    .config("spark.jars.packages", "org.mongodb.spark:mongo-spark-connector_2.12:3.0.1") \  # Asegúrate de que la versión del conector sea la correcta
    .getOrCreate()

# --- CARGAR LOS DATOS DESDE MONGODB CON SPARK ---
df = spark.read.format("mongodb").load()

# Mostrar el DataFrame cargado desde MongoDB
df.show()

# --- EJEMPLO DE ANÁLISIS CON PYSPARK ---

# Ejemplo: Agrupar los libros por categoría y contar el número de libros por categoría
df_grouped = df.groupBy("categoria").agg(F.count("*").alias("count"))
df_grouped.show()

# Otro análisis de ejemplo: Mostrar los libros con el valor más alto
df_highest_value = df.orderBy(F.col("valor").desc()).limit(5)
df_highest_value.show()

# --- GUARDAR EL RESULTADO EN UN ARCHIVO (opcional) ---
# Puedes guardar el resultado de tu análisis en formato Parquet o CSV
df_grouped.write.format("parquet").save("/path/to/save/resultado_analisis")
```
