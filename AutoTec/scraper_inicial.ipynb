```python
import requests
from bs4 import BeautifulSoup

# URL de la página que quieres scrapear
url = "https://www.aspillagahornauer.cl/?utm_campaign=gmb&utm_medium=organic&utm_source=local" 

# Realiza la petición a la página
respuesta = requests.get(url)

# Parsear el contenido HTML
sopa = BeautifulSoup(respuesta.text, 'html.parser')

# Buscar el precio usando la clase 'price-value'
precios = sopa.find_all('span', class_='price-value')

# Mostrar los precios encontrados
for precio in precios:
    print(f"Precio encontrado: {precio.text.strip()}")
```
