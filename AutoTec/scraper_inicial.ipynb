```python
import requests 
from bs4 import BeautifulSoup 
import re

# Configuración de entrada 
url = "https://www.aspillagahornauer.cl/" 
headers = {"User-Agent": "Mozilla/5.0"} 

# 1. Petición al servidor 
respuesta = requests.get(url, headers=headers) 

# 2. Convertir el HTML en un objeto de Python 
sopa = BeautifulSoup(respuesta.text, 'html.parser') 

# 3. Buscar textos
resultados = sopa.find_all(string=True)

# 4. Mostrar solo precios bien formateados
for dato in resultados:
    texto = dato.strip()
    
    # 🔥 FILTRO MEJORADO (solo precios tipo $1.000.000)
    if re.fullmatch(r"\$\s?\d{1,3}(\.\d{3})+", texto):
        print(f"Precio encontrado: {texto}")
```
