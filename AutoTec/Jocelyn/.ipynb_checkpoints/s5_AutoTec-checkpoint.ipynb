```python
import os
import re
import time
from pymongo import MongoClient
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Cierra procesos viejos que hayan quedado abiertos
os.system("pkill -9 chrome")
os.system("pkill -9 chromedriver")
os.system("rm -rf /tmp/.com.google.Chrome.*")
os.system("rm -rf /tmp/.org.chromium.Chromium.*")
print("🧹 Limpieza de procesos y temporales completada.")

# --- VARIABLES GENERALES ---
NOMBRE_GRUPO = "AutoTec"
datos_finales = []   # Se define fuera del try para que siempre exista
driver = None        # Se define fuera del try para poder cerrarlo con seguridad

# --- PASO 1: CONFIGURACIÓN DEL NAVEGADOR ---
options = Options()
options.binary_location = "/usr/bin/google-chrome"  # Ruta del binario de Chrome

# Argumentos de estabilidad para Docker
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--disable-gpu")
options.add_argument("--disable-software-rasterizer")
options.add_argument("--window-size=1920,1080")
options.add_argument("--remote-debugging-port=9222")

# User-Agent común
options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
    "AppleWebKit/537.36 (KHTML, like Gecko) "
    "Chrome/120.0.0.0 Safari/537.36"
)

# --- FUNCIONES DE APOYO ---
def limpiar_precio(texto):
    if not texto:
        return 0.0
    v_limpio = re.sub(r"[^\d]", "", str(texto))
    return float(v_limpio) if v_limpio.isdigit() else 0.0

def extraer_precio(texto):
    # Ejemplo: $10.790.000
    match = re.search(r"\$\s?[\d\.]+", texto)
    return match.group(0) if match else None

def extraer_anio(texto):
    # Ejemplo: 2022
    match = re.search(r"\b(19\d{2}|20\d{2})\b", texto)
    return match.group(0) if match else None

def extraer_kilometraje(texto):
    # Ejemplo: 43.237 Kms o 43124 KM
    match = re.search(r"(\d[\d\.\,]*)\s*(Kms|KM|kms|km)", texto)
    return match.group(0) if match else None

def extraer_combustible(texto):
    # En tu enunciado lo llamas bencina, pero en Trovit aparece como combustible
    opciones = ["Gasolina", "Diesel", "Diésel", "Híbrido", "Eléctrico", "Gas"]
    for op in opciones:
        if op.lower() in texto.lower():
            return op
    return None

def extraer_motor(texto):
    # En estos avisos el "motor" muchas veces viene como transmisión/tipo:
    # Mecánica / Automática
    opciones = ["Mecánica", "Automática", "Manual"]
    for op in opciones:
        if op.lower() in texto.lower():
            return op
    return None

def extraer_marca(nombre):
    # Como tu búsqueda es Chevrolet, la marca puede fijarse o extraerse del nombre
    if nombre and "chevrolet" in nombre.lower():
        return "Chevrolet"
    return "Chevrolet"

def extraer_modelo(nombre):
    # Quita la marca y deja el resto como modelo
    if not nombre:
        return None
    modelo = re.sub(r"(?i)^chevrolet\s+", "", nombre).strip()
    return modelo if modelo else None

try:
    # Inicia el navegador
    driver = webdriver.Chrome(options=options)
    print(" Navegador iniciado correctamente.")

    # --- PASO 2: NAVEGACIÓN Y EXTRACCIÓN ---
    limite_paginas = 1
    driver.get("https://autos.cari.cl/autos/hyundai")

    for nivel_pagina in range(limite_paginas):
        print(f"--- Procesando Página {nivel_pagina + 1} ---")

        # Espera fija para que la página termine de cargar
        time.sleep(8)

        # Espera explícita a que existan resultados
        WebDriverWait(driver, 20).until(
            EC.presence_of_all_elements_located(
                (By.CSS_SELECTOR, "article, li[itemtype='http://schema.org/Product'], div[class*='item']")
            )
        )

        bloques = driver.find_elements(
            By.CSS_SELECTOR,
            "article, li[itemtype='http://schema.org/Product'], div[class*='item']"
        )

        print(f" Bloques encontrados: {len(bloques)}")

        for bloque in bloques:
            try:
                texto = bloque.text.strip()
                if not texto:
                    continue

                lineas = [x.strip() for x in texto.split("\n") if x.strip()]
                if not lineas:
                    continue

                # Tomamos la primera línea útil como nombre principal
                nombre = lineas[0]

                # Filtro para evitar bloques basura
                if len(nombre) < 5:
                    continue

                precio = extraer_precio(texto)
                anio = extraer_anio(texto)
                kilometraje = extraer_kilometraje(texto)
                combustible = extraer_combustible(texto)
                motor = extraer_motor(texto)
                marca = extraer_marca(nombre)
                modelo = extraer_modelo(nombre)

                # Guardamos aunque falte alguno, pero debe tener al menos nombre y precio
                if nombre and precio:
                    datos_finales.append({
                        "identificador": nombre,
                        "marca": marca,
                        "modelo": modelo,
                        "kilometraje": kilometraje,
                        "anio": anio,
                        "bencina": combustible,
                        "motor": motor,
                        "valor": precio,
                        "fecha_captura": time.strftime("%Y-%m-%d %H:%M:%S"),
                        "grupo": NOMBRE_GRUPO
                    })
            except:
                continue

        # Intenta avanzar a la siguiente página
        try:
            btn_sig = driver.find_element(By.PARTIAL_LINK_TEXT, "Siguiente")
            driver.execute_script("arguments[0].click();", btn_sig)
            time.sleep(5)
        except:
            print("No se encontró el botón siguiente o ya es la última página.")
            break

    print(f" Extracción terminada: {len(datos_finales)} productos.")

except Exception as e:
    print(f" Error en Selenium: {e}")

finally:
    # Cierra el navegador solo si logró abrirse
    if driver is not None:
        try:
            driver.quit()
        except:
            pass

# --- PASO 3: GUARDAR EN MONGODB ---
# --- PASO 3: GUARDAR EN MONGODB ATLAS ---
try:
    client = MongoClient("mongodb", 27017, serverSelectionTimeoutMS=5000)
    db = client["proyecto_bigdata"]
    coleccion = db["datos_scraping"]

    client.admin.command("ping")
    print("Conexión exitosa a MongoDB")

    if datos_finales:
        for d in datos_finales:
            v_limpio = str(d["valor"]).replace("$", "").replace(".", "").replace(",", "").strip()
            d["valor"] = float(v_limpio) if v_limpio.isdigit() else 0.0

        coleccion.insert_many(datos_finales)
        print("Datos cargados en MongoDB correctamente.")
    else:
        print("No hay datos para guardar.")

except Exception as e:
    print(f"Error en MongoDB: {e}")
```
