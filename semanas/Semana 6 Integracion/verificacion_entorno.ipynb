```python
import os
import time
from pymongo import MongoClient
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service

# --- PASO 0: VARIABLES DE ENTORNO ---
# Esto es vital para que Chrome use el "monitor virtual" que configuraste en Docker
os.environ["DISPLAY"] = ":99"
os.environ["XDG_RUNTIME_DIR"] = "/tmp/runtime-root"

# --- PASO 1: CONFIGURACIÓN DE CHROME ---
options = Options()
options.binary_location = "/usr/bin/google-chrome"
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--disable-gpu")
options.add_argument("--disable-software-rasterizer")
options.add_argument("--window-size=1366,768")
options.add_argument("--remote-debugging-port=9222")
options.add_argument("--headless=new") # Prueba primero con headless dentro de Docker

print("🧪 Iniciando test de conectividad...")

# --- PASO 2: TEST DE NAVEGADOR (GOOGLE) ---
try:
    print("🚀 Arrancando Chrome...")
    driver = webdriver.Chrome(service=Service(), options=options)
    
    print("🌐 Navegando a Google para verificar...")
    driver.get("https://www.google.com")
    
    titulo = driver.title
    print(f"✅ ¡Conexión exitosa! El título de la página es: {titulo}")
    
    # Si llegamos aquí, Selenium está perfecto.
    driver.quit()
    print("🛑 Navegador cerrado.")

except Exception as e:
    print(f"❌ Error en Selenium: {e}")

# --- PASO 3: TEST DE MONGODB (OPCIONAL PERO RECOMENDADO) ---
print("\n🍃 Verificando conexión a MongoDB...")
try:
    # Cambia 'mongodb' por el nombre de tu servicio en docker-compose si es distinto
    client = MongoClient("mongodb://mongodb:27017/", serverSelectionTimeoutMS=2000)
    client.server_info() # Esto fuerza una conexión real
    print("✅ ¡Conexión exitosa a MongoDB!")
except Exception as e:
    print(f"❌ No se pudo conectar a MongoDB: {e}")

print("\n--- Test Finalizado ---")
```
