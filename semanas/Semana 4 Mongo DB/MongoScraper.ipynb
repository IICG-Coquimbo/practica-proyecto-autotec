# Conectar el Scraper a la BD y guardar
Se usa el script de la semana 3

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
from pymongo import MongoClient

NOMBRE_GRUPO = "AutoTec"

options = Options()
options.add_argument("--headless=new")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920,1080")

driver = webdriver.Chrome(options=options)

datos_finales = []

try:
    driver.get("https://autos.cari.cl/")
    time.sleep(5)

    print("Título:", driver.title)

    enlaces = driver.find_elements(By.TAG_NAME, "a")

    for enlace in enlaces:
        try:
            href = enlace.get_attribute("href")
            texto = enlace.text

            if href and "auto" in href.lower():
                datos_finales.append({
                    "identificador": texto,
                    "valor": href,
                    "fecha_captura": time.strftime("%Y-%m-%d %H:%M:%S"),
                    "grupo": NOMBRE_GRUPO
                })
        except:
            continue

    print("Registros encontrados:", len(datos_finales))

except Exception as e:
    print("Fallo en la arquitectura de scraping:", e)

finally:
    driver.quit()


# MongoDB
client = MongoClient('database', 27017)
db = client['proyecto_bigdata']
coleccion = db['datos_scraping']

for dato in datos_finales:
    coleccion.insert_one(dato)

print("Datos insertados:", len(datos_finales))
```
