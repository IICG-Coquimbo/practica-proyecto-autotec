```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

# --- CONFIGURACIÓN DE SPARK PARA CARGAR DATOS DESDE MONGODB ---
spark = SparkSession.builder \
    .appName("Analisis_Categorias") \
    .config("spark.mongodb.read.connection.uri", "mongodb://database:27017/proyecto_bigdata.datos_libros") \
    .config("spark.jars.packages", "org.mongodb.spark:mongo-spark-connector_2.12:3.0.1") \
    .getOrCreate()

# Cargar los datos desde MongoDB usando Spark
df = spark.read.format("mongodb").load()

# Muestra el DataFrame
df.show()

# Realizar análisis utilizando PySpark (ejemplo)
df_grouped = df.groupBy("categoria").agg(F.count("*").alias("count"))
df_grouped.show()

# Ejemplo de análisis adicional: Obtener los libros más caros
df_highest_value = df.orderBy(F.col("valor").desc()).limit(5)
df_highest_value.show()

# --- GUARDAR EL RESULTADO EN UN ARCHIVO (opcional) ---
# Puedes guardar el resultado de tu análisis en formato Parquet o CSV
df_grouped.write.format("parquet").save("/path/to/save/resultado_analisis")
```
