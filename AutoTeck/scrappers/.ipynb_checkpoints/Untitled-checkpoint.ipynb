```python
import os
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from pymongo import MongoClient

def ejecutar_extraccion():
    datos_finales = []  # Lista donde se guardarán los datos extraídos

    # 🌟 Configuración de Selenium
    options = Options()
    options.add_argument("--headless")  # Ejecución sin abrir ventana del navegador 
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")

    # Iniciar el driver de Chrome
    driver = webdriver.Chrome(options=options)

    try:
        driver.get("http://books.toscrape.com/")  # 🌍 URL a scrapear
        
        limite_paginas = 3  # Definir la cantidad de páginas a scrapear

        for nivel_pagina in range(limite_paginas):
            print(f"🛠️ --- Procesando Página {nivel_pagina + 1} ---")

            # 🕒 Espera a que carguen los resultados
            WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.CLASS_NAME, "product_pod"))
            )

            bloques_primer_nivel = driver.find_elements(By.CLASS_NAME, "product_pod")
            
            for bloque in bloques_primer_nivel:
                try:
                    # 📖 Extraer nombre y precio de cada producto
                    nombre = bloque.find_element(By.TAG_NAME, "h3").find_element(By.TAG_NAME, "a").get_attribute("title")
                    precio = bloque.find_element(By.CLASS_NAME, "price_color").text

                    # Guardar los datos extraídos en la lista
                    datos_finales.append({
                        "identificador": nombre,
                        "valor": precio,
                        "fecha_captura": time.strftime("%Y-%m-%d %H:%M:%S"),
                        "grupo": "AutoTec"  # 👨‍💻 Nombre del grupo
                    })
                    print(f"✅ Producto '{nombre}' añadido a la lista.")

                except Exception as e:
                    print(f"⚠️ Error al procesar un bloque: {e}")

            # 🏃 Intentar pasar a la siguiente página
            try:
                disparador_siguiente = driver.find_element(By.CSS_SELECTOR, "li.next a")
                disparador_siguiente.click()
                time.sleep(5)
                print("➡️ Avanzando a la siguiente página...")
            except:
                print("❌ No se encontró el botón siguiente o ya es la última página.")
                break

        print(f"🎉 Extracción terminada: {len(datos_finales)} productos extraídos.")
        
    except Exception as e:
        print(f"💥 Error en Selenium: {e}")
        
    finally:
        driver.quit()

    # 🔥 MONGO: Conectamos y guardamos los datos en MongoDB
    try:
        client = MongoClient("mongodb+srv://Martin_Rojas:Ovalle13@cluster0.kxe0cns.mongodb.net/AutoTec_db?retryWrites=true&w=majority")
        db = client["AutoTec_db"]
        coleccion = db["AutoTec"]  # 🏆 Usa el nombre de la colección que prefieras

        if datos_finales:
            coleccion.insert_many(datos_finales)
            print("💾 Datos guardados en MongoDB correctamente. ¡Éxito! 🎉")
        else:
            print("⚠️ No hay datos para guardar.")

    except Exception as e:
        print(f"❌ Error en MongoDB: {e}")

    return datos_finales  # 🚀 Retornar los datos extraídos para visualización
```

```python
import os
import time
import re
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# --- LIMPIEZA ---
os.system("taskkill /F /IM chrome.exe")  # En Windows, usar taskkill para cerrar Chrome
os.system("taskkill /F /IM chromedriver.exe")  # Cerrar cualquier instancia de Chromedriver
print("🧹 Limpieza de procesos y temporales completada.")

def ejecutar_extraccion():
    NOMBRE_GRUPO = "AutoTec"
    USUARIO = "Martin"  # O cualquier nombre que prefieras
    URL_OBJETIVO = "https://seminuevos.aspillagahornauer.cl/stock-seminuevos/?gad_source=1&gad_campaignid=23774006351&gbraid=0AAAAApjXRfxXdyCpHu2DmvhtmvSR6MsDx&gclid=CjwKCAjwqazPBhALEiwAOuXqdJdZD-MPeeY6b_7MYI01cmCua8Y90RnjwcV0ZlpiVec0He3abn3_IxoCcvIQAvD_BwE"
    datos_finales = []

    # --- CONFIGURACIÓN DE SELENIUM ---
    options = Options()
    options.add_argument("--headless")  # Ejecutar sin interfaz gráfica
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--disable-gpu")
    options.add_argument("--window-size=1920,1080")
    options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36")

    try:
        driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
        print("🚀 Navegador iniciado correctamente.")

        # --- EJECUTAR SCRAPING ---
        driver.get(URL_OBJETIVO)

        # Esperar que los elementos carguen
        WebDriverWait(driver, 20).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".card-body"))
        )

        limite_paginas = 3  # Cambia este número si quieres más páginas

        for nivel_pagina in range(limite_paginas):
            print(f"--- Procesando Página {nivel_pagina + 1} ---")

            # Hacer scroll para cargar más autos
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(5)

            # Obtener todos los vehículos
            vehiculos = driver.find_elements(By.CSS_SELECTOR, ".card-body")

            for vehiculo in vehiculos:
                try:
                    # Extraer nombre y modelo
                    nombre = vehiculo.find_element(By.CSS_SELECTOR, "a.rmv_txt_drctn").text

                    # Extraer marca
                    marca = vehiculo.find_element(By.CSS_SELECTOR, "div.value").text

                    # Extraer precio
                    precio = vehiculo.find_element(By.CSS_SELECTOR, "span.heading-font").text
                    precio = re.sub(r'[^0-9]', '', precio)  # Limpieza de precio

                    # Extraer kilometraje
                    kilometraje = vehiculo.find_element(By.XPATH, "//div[contains(text(), 'kms')]").text

                    # Extraer año
                    anio = vehiculo.find_element(By.XPATH, "//div[contains(text(), '202')]").text

                    # Guardar los datos extraídos
                    datos_finales.append({
                        "marca": marca.strip(),
                        "modelo": nombre.strip(),
                        "precio": float(precio),
                        "kilometraje": kilometraje.strip(),
                        "anio": anio.strip(),
                        "fecha_captura": time.strftime("%Y-%m-%d %H:%M:%S"),
                        "grupo": NOMBRE_GRUPO,
                        "usuario": USUARIO
                    })

                except Exception as e:
                    print(f"Error al procesar un vehículo: {e}")
                    continue

            # Intentar pasar a la siguiente página
            try:
                siguiente_pagina = driver.find_element(By.CSS_SELECTOR, "a.pagination__next")
                siguiente_pagina.click()
                time.sleep(5)
                print(f"👉 Avanzando a la página {nivel_pagina + 2}...")

            except Exception:
                print("No se pudo avanzar más: ya es la última página o el botón cambió.")
                break

        print(f"✅ Extracción terminada: {len(datos_finales)} autos.")
        return datos_finales

    except Exception as e:
        print(f"❌ Error en Selenium: {e}")
        return []

    finally:
        if driver is not None:
            driver.quit()

# --- GUARDAR EN MONGODB ---
def guardar_en_mongodb(datos):
    from pymongo import MongoClient

    try:
        client = MongoClient("mongodb+srv://neiel_cortes:neiel0330@cluster0.eo0kyfv.mongodb.net/AutoTec_db?retryWrites=true&w=majority")
        db = client["AutoTec_db"]
        coleccion = db["seminuevos"]

        if datos:
            coleccion.insert_many(datos)
            print(f"💾 {len(datos)} autos guardados en MongoDB.")
        else:
            print("⚠️ No hay datos para guardar.")
    except Exception as e:
        print(f"❌ Error en MongoDB: {e}")

# Ejecutar el scraper y guardar en MongoDB
datos_extraidos = ejecutar_extraccion()
guardar_en_mongodb(datos_extraidos)
```
