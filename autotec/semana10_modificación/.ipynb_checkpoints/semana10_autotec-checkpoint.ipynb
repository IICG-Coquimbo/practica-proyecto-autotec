```python
# ============================================
# SEMANA 10 - CLUSTERING AUTOTEC
# K-Means + PCA + DBSCAN
# Variables: marca, modelo, year, kilometraje,
#            combustible, ciudad, precio
# ============================================

from pymongo import MongoClient
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, DBSCAN

# ============================================
# CONEXION Y CARGA DE DATOS
# ============================================

uri = "mongodb+srv://neiel_cortes:neiel0330@cluster0.eo0kyfv.mongodb.net/"
client = MongoClient(uri)
db = client["proyecto_bigdata"]
collection = db["Contenedor_Autos_Limpio"]

data = list(collection.find())
df = pd.DataFrame(data)

print(f"Total de registros: {len(df)}")
print(f"Columnas: {list(df.columns)}")
df.head()
```

```python
# ============================================
# PREPARACION DE DATOS
# Convertir variables categoricas a numericas
# ============================================

# Seleccionamos solo las 7 variables del proyecto
df_trabajo = df[["marca", "modelo", "year", "kilometraje", "combustible", "ciudad", "precio"]].copy()

# Eliminar nulos
df_trabajo = df_trabajo.dropna()

print(f"Registros despues de limpiar nulos: {len(df_trabajo)}")

# Codificar variables categoricas con LabelEncoder
le_marca       = LabelEncoder()
le_modelo      = LabelEncoder()
le_combustible = LabelEncoder()
le_ciudad      = LabelEncoder()

df_trabajo["marca_cat"]       = le_marca.fit_transform(df_trabajo["marca"].astype(str))
df_trabajo["modelo_cat"]      = le_modelo.fit_transform(df_trabajo["modelo"].astype(str))
df_trabajo["combustible_cat"] = le_combustible.fit_transform(df_trabajo["combustible"].astype(str))
df_trabajo["ciudad_cat"]      = le_ciudad.fit_transform(df_trabajo["ciudad"].astype(str))

# DataFrame solo numerico para clustering
df_clustering = df_trabajo[[
    "marca_cat",
    "modelo_cat",
    "year",
    "kilometraje",
    "combustible_cat",
    "ciudad_cat",
    "precio"
]].copy()

df_clustering["year"]        = df_clustering["year"].astype(float)
df_clustering["kilometraje"] = df_clustering["kilometraje"].astype(float)
df_clustering["precio"]      = df_clustering["precio"].astype(float)

print("\nEsquema del DataFrame de clustering:")
print(df_clustering.dtypes)
df_clustering.head()
```

```python
# ============================================
# PASO 1: PCA
# Reduccion a 2 componentes para visualizacion
# ============================================

# 1. Escalamos los datos (K-Means es sensible a la escala)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_clustering)

# 2. PCA para reducir a 2 componentes
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Matriz de influencia (Loadings)
variables = ["marca_cat", "modelo_cat", "year", "kilometraje", "combustible_cat", "ciudad_cat", "precio"]
pc_loadings = pd.DataFrame(
    pca.components_.T,
    columns=["PC1", "PC2"],
    index=variables
)
print("Matriz de Influencia (Loadings):")
print(pc_loadings)

# Varianza explicada
print(f"\nVarianza explicada PC1: {round(pca.explained_variance_ratio_[0]*100, 2)}%")
print(f"Varianza explicada PC2: {round(pca.explained_variance_ratio_[1]*100, 2)}%")

# Grafico PCA sin clusters
pdf_pca = pd.DataFrame(X_pca, columns=["PC1", "PC2"])
pdf_pca["marca"] = df_trabajo["marca_cat"].values

pc1_min, pc1_max = np.percentile(pdf_pca["PC1"], [1, 99])
pc2_min, pc2_max = np.percentile(pdf_pca["PC2"], [1, 99])

plt.figure(figsize=(11, 7))
sns.set_style("darkgrid")
sns.scatterplot(
    data=pdf_pca, x="PC1", y="PC2",
    color="teal", s=60, alpha=0.5, edgecolor=None
)
plt.xlim(pc1_min - 1, pc1_max + 1)
plt.ylim(pc2_min - 1, pc2_max + 1)
plt.title("Distribucion Espacial de Autos Usados (PCA)", fontsize=15)
plt.xlabel("Componente Principal 1", fontsize=12)
plt.ylabel("Componente Principal 2", fontsize=12)
plt.tight_layout()
plt.savefig("pca_distribucion.png", dpi=150)
plt.show()
print("Grafico guardado: pca_distribucion.png")
```

```python
# ============================================
# PASO 2: METODO DEL CODO
# Encontrar el k optimo para K-Means
# ============================================

cost = []
K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    cost.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K_range, cost, "bx-")
plt.xlabel("Numero de Clusters (k)")
plt.ylabel("Costo (Inercia)")
plt.title("Metodo del Codo para Autos Usados AutoTec")
plt.grid(linestyle="--", alpha=0.5)
plt.tight_layout()
plt.savefig("metodo_codo.png", dpi=150)
plt.show()
print("Grafico guardado: metodo_codo.png")
```

```python
# ============================================
# PASO 3: CLUSTERING FINAL K-MEANS
# ============================================

k_optimo = 3
kmeans_final = KMeans(n_clusters=k_optimo, random_state=42, n_init=10)
clusters_kmeans = kmeans_final.fit_predict(X_scaled)

# Agregar prediccion al DataFrame
df_trabajo["cluster_kmeans"] = clusters_kmeans

# Resumen por cluster
print(f"Clustering K-Means con k={k_optimo}:")
print(df_trabajo.groupby("cluster_kmeans")[["precio", "kilometraje", "year"]].mean().round(0))
```

```python
# ============================================
# PASO 4: VISUALIZACION K-MEANS EN PCA
# ============================================

pdf_visual = pd.DataFrame(X_pca, columns=["PC1", "PC2"])
pdf_visual["prediction"] = clusters_kmeans
pdf_visual["marca"]      = df_trabajo["marca"].values

pc1_min, pc1_max = np.percentile(pdf_visual["PC1"], [1, 99])
pc2_min, pc2_max = np.percentile(pdf_visual["PC2"], [1, 99])

plt.figure(figsize=(12, 8))
sns.set_style("darkgrid")
sns.scatterplot(
    data=pdf_visual, x="PC1", y="PC2",
    hue="prediction", palette="viridis",
    s=80, alpha=0.7, edgecolor=None
)
plt.xlim(pc1_min - 1, pc1_max + 1)
plt.ylim(pc2_min - 1, pc2_max + 1)
plt.title(f"Visualizacion de Clusters en Espacio PCA (k={k_optimo})", fontsize=15)
plt.xlabel("Componente Principal 1", fontsize=12)
plt.ylabel("Componente Principal 2", fontsize=12)
plt.legend(title="Cluster", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.savefig("kmeans_clusters.png", dpi=150)
plt.show()
print("Grafico guardado: kmeans_clusters.png")
```

```python
# ============================================
# PASO 5: DBSCAN
# ============================================

dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters_dbscan = dbscan.fit_predict(X_pca)

pdf_dbscan = pd.DataFrame(X_pca, columns=["PC1", "PC2"])
pdf_dbscan["cluster_dbscan"] = clusters_dbscan

n_clusters = len(set(clusters_dbscan)) - (1 if -1 in clusters_dbscan else 0)
n_ruido    = list(clusters_dbscan).count(-1)
print(f"Clusters encontrados por DBSCAN: {n_clusters}")
print(f"Puntos de ruido: {n_ruido}")

pc1_min, pc1_max = np.percentile(pdf_dbscan["PC1"], [1, 99])
pc2_min, pc2_max = np.percentile(pdf_dbscan["PC2"], [1, 99])

plt.figure(figsize=(12, 8))
sns.set_style("darkgrid")
sns.scatterplot(
    data=pdf_dbscan, x="PC1", y="PC2",
    hue="cluster_dbscan", palette="Set1",
    s=70, alpha=0.6
)
plt.xlim(pc1_min - 1, pc1_max + 1)
plt.ylim(pc2_min - 1, pc2_max + 1)
plt.title("Clustering con DBSCAN (Deteccion de Densidad y Ruido)", fontsize=16)
plt.legend(title="Cluster (-1 = Ruido)", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.savefig("dbscan_clusters.png", dpi=150)
plt.show()
print("Grafico guardado: dbscan_clusters.png")

print("\nClustering completado correctamente.")
```
