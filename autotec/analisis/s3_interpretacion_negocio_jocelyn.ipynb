# Interpretación de Resultados EDA

## Objetivo del análisis

El análisis exploratorio de datos (EDA) tuvo como propósito comprender el comportamiento de las principales variables asociadas al mercado de vehículos usados presentes en la base de datos AutoTec.

## Hallazgos principales

A partir del análisis realizado, se identificó que el precio de los vehículos mantiene una relación directa con el año de fabricación y una relación inversa con el kilometraje. Esto significa que los vehículos más nuevos tienden a presentar precios más altos, mientras que aquellos con mayor nivel de uso suelen disminuir su valor comercial.

El análisis de correlación confirmó que las variables año y kilometraje poseen capacidad explicativa sobre el comportamiento del precio, siendo factores relevantes para comprender la depreciación de los automóviles usados.

Asimismo, la detección de valores atípicos permitió identificar vehículos cuyos precios se encuentran considerablemente por encima del promedio del conjunto de datos. Estos registros pueden corresponder a vehículos premium, modelos exclusivos o automóviles con características diferenciadoras.

## Conclusión

Los resultados obtenidos permiten concluir que las variables precio, kilometraje, año y marca representan factores relevantes para el análisis del mercado automotriz usado. Además, estos hallazgos proporcionan una base sólida para futuras etapas de segmentación mediante clustering y construcción de modelos predictivos orientados a la estimación del precio de vehículos.
