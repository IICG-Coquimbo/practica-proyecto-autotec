```python
# --- PASO 0: LIMPIEZA TOTAL Y REPARACIÓN ---
import os
import time
from pymongo import MongoClient
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Cierra procesos viejos que hayan quedado abiertos
os.system("pkill -9 chrome")
os.system("pkill -9 chromedriver")
os.system("rm -rf /tmp/.com.google.Chrome.*")
os.system("rm -rf /tmp/.org.chromium.Chromium.*")
print("🧹 Limpieza de procesos y temporales completada.")

# --- VARIABLES GENERALES ---
NOMBRE_GRUPO = "AutoTec"
USUARIO = "daniela"
lista_autos = []
driver = None

# --- PASO 1: CONFIGURACIÓN DEL NAVEGADOR ---
options = Options()
options.binary_location = "/usr/bin/google-chrome"

# Argumentos de estabilidad para Docker
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--disable-gpu")
options.add_argument("--disable-software-rasterizer")
options.add_argument("--window-size=1920,1080")
options.add_argument("--remote-debugging-port=9222")
options.add_argument("--headless=new")

# User-Agent común
options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
    "AppleWebKit/537.36 (KHTML, like Gecko) "
    "Chrome/120.0.0.0 Safari/537.36"
)

try:
    # Inicia el navegador
    driver = webdriver.Chrome(options=options)
    print("✅ Navegador iniciado correctamente.")

    # --- PASO 2: NAVEGACIÓN Y EXTRACCIÓN ---
    limite_paginas = 3
    driver.get("https://www.yapo.cl/autos")

    for nivel_pagina in range(limite_paginas):
        print(f"--- Procesando Página {nivel_pagina + 1} ---")

        # Espera fija para que la página termine de cargar
        time.sleep(10)

        # Espera explícita a que existan resultados
        WebDriverWait(driver, 20).until(
            EC.presence_of_all_elements_located(
                (By.CSS_SELECTOR, "a.d3-ad-tile__description")
            )
        )

        bloques = driver.find_elements(
            By.CSS_SELECTOR, "a.d3-ad-tile__description"
        )

        print(f"Autos encontrados en página {nivel_pagina + 1}: {len(bloques)}")

        for bloque in bloques:
            try:
                nombre = bloque.find_element(
                    By.CSS_SELECTOR, "span.d3-ad-tile__title"
                ).text

                try:
                    precio = bloque.find_element(
                        By.CSS_SELECTOR, "div.d3-ad-tile__price"
                    ).text
                except:
                    precio = "0"

                try:
                    ubicacion = bloque.find_element(
                        By.CSS_SELECTOR, "div.d3-ad-tile__location"
                    ).text
                except:
                    ubicacion = ""

                link = bloque.get_attribute("href")

                lista_autos.append({
                    "identificador": nombre,
                    "valor": precio,
                    "ubicacion": ubicacion,
                    "url": link,
                    "fecha_captura": time.strftime("%Y-%m-%d %H:%M:%S"),
                    "grupo": NOMBRE_GRUPO,
                    "usuario": USUARIO
                })
            except:
                continue

        # Intenta avanzar a la siguiente página
        try:
            links = driver.find_elements(By.TAG_NAME, "a")
            btn_sig = None

            for link in links:
                texto = (link.text or "").strip().lower()
                aria = (link.get_attribute("aria-label") or "").strip().lower()
                href = (link.get_attribute("href") or "").strip().lower()

                if "siguiente" in texto or "siguiente" in aria:
                    btn_sig = link
                    break

                if "page=" in href and str(nivel_pagina + 2) in href:
                    btn_sig = link
                    break

            if btn_sig:
                driver.execute_script("arguments[0].click();", btn_sig)
                time.sleep(5)
                print("✅ Avanzó a la siguiente página.")
            else:
                print("No se encontró el botón siguiente o ya es la última página.")
                break

        except Exception as e:
            print(f"No se pudo avanzar de página: {e}")
            break

    print(f"✅ Extracción terminada: {len(lista_autos)} autos.")

except Exception as e:
    print(f"❌ Error en Selenium: {e}")

finally:
    # Cierra el navegador solo si logró abrirse
    if driver is not None:
        try:
            driver.quit()
        except:
            pass

# --- PASO 3: GUARDAR EN MONGODB ---
try:
    client = MongoClient("mongodb", 27017, serverSelectionTimeoutMS=5000)
    db = client["proyecto_bigdata"]
    coleccion = db["YapoAutos"]

    if lista_autos:
        for d in lista_autos:
            # Limpia el valor antes de convertirlo
            v_limpio = (
                str(d["valor"])
                .replace("$", "")
                .replace(".", "")
                .replace(",", "")
                .strip()
            )
            d["valor"] = float(v_limpio) if v_limpio.isdigit() else 0.0

        coleccion.insert_many(lista_autos)
        print("✅ Datos cargados en MongoDB correctamente.")
    else:
        print("⚠️ No hay datos para guardar.")

except Exception as e:
    print(f"❌ Error en MongoDB: {e}")
```

```python
def ejecutar_extraccion():

    import os
    import time
    from pymongo import MongoClient
    from selenium import webdriver
    from selenium.webdriver.chrome.options import Options
    from selenium.webdriver.common.by import By
    from selenium.webdriver.support.ui import WebDriverWait
    from selenium.webdriver.support import expected_conditions as EC

    os.system("pkill -9 chrome")
    os.system("pkill -9 chromedriver")

    lista_autos = []
    driver = None

    NOMBRE_GRUPO = "autotec"
    USUARIO = "daniela"

    options = Options()
    options.binary_location = "/usr/bin/google-chrome"
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--headless=new")

    try:
        driver = webdriver.Chrome(options=options)

        for pagina in range(5):

            url = "https://www.yapo.cl/autos-usados" if pagina == 0 else f"https://www.yapo.cl/autos-usados.{pagina+1}"
            driver.get(url)

            time.sleep(8)

            bloques = WebDriverWait(driver, 20).until(
                EC.presence_of_all_elements_located(
                    (By.CSS_SELECTOR, "a.d3-ad-tile__description")
                )
            )

            for bloque in bloques:
                try:
                    # --- NOMBRE ---
                    nombre = bloque.find_element(
                        By.CSS_SELECTOR, "span.d3-ad-tile__title"
                    ).text.lower()

                    partes = nombre.split()
                    marca = partes[0] if len(partes) > 0 else ""
                    modelo = " ".join(partes[1:]) if len(partes) > 1 else ""

                    # --- PRECIO ---
                    try:
                        precio = bloque.find_element(
                            By.CSS_SELECTOR, "div.d3-ad-tile__price"
                        ).text
                    except:
                        precio = "0"

                    precio_limpio = (
                        precio.replace("$", "")
                        .replace(".", "")
                        .replace(",", "")
                        .strip()
                    )
                    precio = float(precio_limpio) if precio_limpio.isdigit() else 0.0

                    # --- CIUDAD ---
                    try:
                        ciudad = bloque.find_element(
                            By.CSS_SELECTOR, "div.d3-ad-tile__location"
                        ).text.lower()
                    except:
                        ciudad = ""

                    # --- URL ---
                    link = bloque.get_attribute("href")

                    # --- DETALLES (AÑO, COMBUSTIBLE, KM) ---
                    detalles = bloque.find_elements(
                        By.CSS_SELECTOR, "li.d3-ad-tile__details-item"
                    )

                    año = ""
                    combustible = ""
                    kilometraje = ""

                    for d in detalles:
                        texto = d.text.lower()

                        if texto.isdigit() and len(texto) == 4:
                            año = texto

                        elif "km" in texto:
                            kilometraje = texto

                        elif any(c in texto for c in ["bencina", "diesel", "diésel", "hibrido", "híbrido", "electrico", "eléctrico"]):
                            combustible = texto

                    # --- GUARDAR ---
                    lista_autos.append({
                        "marca": marca,
                        "modelo": modelo,
                        "año": año,
                        "kilometraje": kilometraje,
                        "combustible": combustible,
                        "ciudad": ciudad,
                        "precio": precio,
                        "url": link,
                        "nombre": nombre,
                        "fecha_captura": time.strftime("%Y-%m-%d %H:%M:%S"),
                        "grupo": NOMBRE_GRUPO,
                        "usuario": USUARIO
                    })

                except Exception as e:
                    print("error:", e)
                    continue

        print(f"autos extraidos: {len(lista_autos)}")

    except Exception as e:
        print("error selenium:", e)

    finally:
        if driver:
            driver.quit()

    # --- MONGO ---
    try:
        client = MongoClient("mongodb", 27017, serverSelectionTimeoutMS=5000)
        db = client["proyecto_bigdata"]
        coleccion = db["yapoautos"]

        if lista_autos:
            coleccion.insert_many(lista_autos)
            print("datos guardados en mongo")

    except Exception as e:
        print("error mongo:", e)

    return lista_autos
```
