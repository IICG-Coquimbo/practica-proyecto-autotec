### Segmentación mediante Clustering con K-Means

Se aplicó el algoritmo **K-Means** sobre las variables `precio`, `kilometraje` y `año`, con un total de **1955 registros limpios**. El objetivo fue identificar segmentos latentes en el mercado automotriz.

**Selección del número de clusters:**
- Se utilizó el **método del codo**, que mostró una clara estabilización de la curva de inercia en k=3 y k=4.  
  Esto indica que a partir de ese punto, agregar más clusters no mejora significativamente la calidad del modelo.
- Se calculó el **Silhouette Score** para cada valor de k:  
  - k=2 = 0.678 (el puntaje más alto, pero demasiado general, ya que solo divide en dos grupos amplios).  
  - k=3 = 0.608 (buena calidad y mayor diferenciación de segmentos).  
  - k=4 = 0.606 (similar a k=3, pero con un grupo adicional que captura autos con características mixtas).  
- A partir de k=5 los puntajes bajan (<0.56), lo que indica pérdida de calidad en la segmentación.  
Por ello, se decidió trabajar con **k=3 o k=4 clusters**, balanceando calidad técnica y utilidad práctica.

**Perfilamiento de clusters:**
- Cluster 0: Autos económicos, alto kilometraje y bajo precio. Segmento de clientes sensibles al costo.  
- Cluster 1: Autos nuevos, marcas premium y precios altos. Segmento de clientes de lujo.  
- Cluster 2: Autos intermedios, kilometraje moderado y precios medios. Segmento de clientes familiares.  
- Cluster 3 (cuando se usa k=4): Autos con características mixtas o valores atípicos. Segmento residual que requiere análisis adicional.

**Conclusión:**  
El algoritmo **K-Means** fue escogido porque permitió una segmentación clara y homogénea del dataset, validada tanto por el método del codo como por el Silhouette Score. La elección de **k=3 o k=4 clusters** asegura que los grupos obtenidos sean técnicamente consistentes y, al mismo tiempo, interpretables en términos de negocio. Esta segmentación constituye un conocimiento explícito que la organización puede utilizar para diseñar estrategias diferenciadas de marketing y ventas.


### Segmentación mediante Clustering con DBSCAN

Se aplicó el algoritmo **DBSCAN** sobre las variables `precio`, `kilometraje` y `año`, utilizando reducción de dimensionalidad con PCA para facilitar la visualización. El dataset final contó con **1955 registros limpios**.

**Selección de parámetros:**
- Se probaron distintas combinaciones de `eps` y `min_samples` para evaluar la calidad del clustering.  
- Los gráficos mostraron cómo variaba la cantidad de clusters y outliers según los parámetros.  
- El **heatmap del Silhouette Score** permitió identificar la mejor configuración.  
- La combinación óptima fue **eps=0.7 y min_samples=3**, con un **Silhouette Score de 0.711**, lo que indica una separación clara entre clusters y ruido.

**Resultados obtenidos:**
- Con estos parámetros, DBSCAN detectó **2 clusters principales** y **4 outliers**.  
- Los outliers fueron resaltados en rojo en la visualización, representando autos con precios o kilometrajes atípicos.  
- La paleta Spectral permitió distinguir claramente los grupos y el ruido.

**Perfilamiento de clusters:**
- Cluster 0: Autos económicos, alto kilometraje y bajo precio.  
- Cluster 1: Autos nuevos, precios altos y marcas premium.  
- Cluster -1: Outliers, que corresponden a vehículos con características extremas o poco representativas del mercado.

**Conclusión:**  
El algoritmo **DBSCAN** fue escogido porque permite detectar tanto clusters como ruido en los datos, algo especialmente útil en un mercado automotriz donde existen vehículos con precios o kilometrajes fuera de lo común. La configuración óptima (eps=0.7, min_samples=3) aseguró una segmentación clara y al mismo tiempo resaltó los autos atípicos, aportando información valiosa para la toma de decisiones de negocio.


### Comparación crítica: K-Means vs DBSCAN

**K-Means**
- Ventajas:  
  - Segmentación clara y homogénea.  
  - Fácil de interpretar y visualizar.  
  - Buen rendimiento con datasets grandes (como los 1955 registros).  
- Limitaciones:  
  - Requiere definir el número de clusters de antemano.  
  - No detecta outliers, los incluye dentro de algún cluster.  
- Justificación en este caso:  
  - El método del codo y el Silhouette Score mostraron que k=3 o k=4 es óptimo.  
  - Los clusters obtenidos son interpretables en términos de negocio (económicos, premium, familiares).  
  - Es ideal cuando se busca segmentación estable y aplicable a estrategias de marketing.

**DBSCAN**
- Ventajas:  
  - No necesita definir el número de clusters previamente.  
  - Detecta automáticamente ruido y outliers.  
  - Útil para datos con distribuciones irregulares.  
- Limitaciones:  
  - Sensible a los parámetros `eps` y `min_samples`.  
  - Puede generar resultados inestables si los datos no tienen densidades claras.  
- Justificación en este caso:  
  - Con eps=0.7 y min_samples=3 se lograron 2 clusters y 4 outliers, con Silhouette Score de 0.711.  
  - Permite identificar autos atípicos (precios o kilometrajes extremos).  
  - La segmentación es menos rica que la de K-Means, ya que solo se formaron dos grupos principales.

**Conclusión crítica y neutra:**  
- K-Means es más adecuado cuando el objetivo es una **segmentación clara y aplicable al negocio**, porque genera clusters homogéneos y fáciles de interpretar.  
- DBSCAN es más adecuado cuando el objetivo es **detectar outliers y ruido**, resaltando vehículos atípicos que podrían distorsionar el análisis.  
- En este proyecto, ambos aportan valor:  
  - **K-Means** se utiliza como algoritmo principal para definir segmentos de mercado (económicos, premium, familiares).  
  - **DBSCAN** se emplea como complemento para identificar autos fuera de rango y enriquecer el análisis.
