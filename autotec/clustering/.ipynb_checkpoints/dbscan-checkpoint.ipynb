```python
# =========================
# LIBRERÍAS
# =========================
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
from pyspark.sql.types import IntegerType, DoubleType
from pyspark.ml.feature import VectorAssembler, StandardScaler, PCA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import DBSCAN
from IPython.display import Markdown, display

# =========================
# CREAR SESIÓN SPARK
# =========================
spark = (
    SparkSession.builder
    .appName("Clustering_DBSCAN_AutoTec")
    .config("spark.jars.packages", "org.mongodb.spark:mongo-spark-connector_2.12:10.1.1")
    .config("spark.mongodb.read.connection.uri", "mongodb+srv://neiel_cortes:neiel0330@cluster0.eo0kyfv.mongodb.net/AutoTec_db")
    .getOrCreate()
)

# =========================
# LEER DATOS
# =========================
df = spark.read.format("mongodb") \
    .option("database", "proyecto_bigdata") \
    .option("collection", "Contenedor_Autos_Limpio") \
    .load()

print("Registros cargados:", df.count())
df.printSchema()

# =========================
# LIMPIEZA Y CASTEO
# =========================
df_cast = df.withColumn("kilometraje", col("kilometraje").cast(IntegerType())) \
            .withColumn("year", col("year").cast(IntegerType())) \
            .withColumn("precio_num", col("precio").cast(DoubleType())) \
            .dropna(subset=["precio_num", "kilometraje", "year"])

# =========================
# VECTORIZACIÓN Y PCA
# =========================
assembler = VectorAssembler(
    inputCols=["precio_num", "kilometraje", "year"],
    outputCol="features"
)
df_vector = assembler.transform(df_cast)

scaler = StandardScaler(inputCol="features", outputCol="scaledFeatures")
scaler_model = scaler.fit(df_vector)
df_scaled = scaler_model.transform(df_vector)

pca = PCA(k=2, inputCol="scaledFeatures", outputCol="pcaFeatures")
pca_model = pca.fit(df_scaled)
df_pca = pca_model.transform(df_scaled)

# =========================
# EXPORTAR A PANDAS
# =========================
pdf_dbscan = df_pca.select("pcaFeatures", "marca").toPandas()
X = np.array(pdf_dbscan["pcaFeatures"].apply(lambda x: x.toArray()).tolist())
pdf_dbscan[["PC1", "PC2"]] = pd.DataFrame(X, columns=["PC1", "PC2"])

# =========================
# DBSCAN
# =========================
eps_value = 0.5
min_samples_value = 5

dbscan = DBSCAN(eps=eps_value, min_samples=min_samples_value)
clusters = dbscan.fit_predict(X)
pdf_dbscan["cluster_dbscan"] = clusters

# =========================
# VISUALIZACIÓN MEJORADA (Spectral)
# =========================
plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")

# Paleta diferenciada automática
palette = sns.color_palette("Spectral", len(set(clusters)))

# Scatter principal con paleta Spectral
sns.scatterplot(
    data=pdf_dbscan,
    x="PC1",
    y="PC2",
    hue="cluster_dbscan",
    palette=palette,
    s=80,
    alpha=0.7,
    edgecolor="black"
)

# Resaltar outliers (-1) en rojo con "x"
outliers = pdf_dbscan[pdf_dbscan["cluster_dbscan"] == -1]
plt.scatter(outliers["PC1"], outliers["PC2"], 
            color="red", s=40, marker="x", label="Ruido (-1)")

# Calcular métricas
num_clusters = len(set(clusters)) - (1 if -1 in clusters else 0)
num_outliers = list(clusters).count(-1)

# Título con parámetros y métricas
plt.title(
    f"Clustering DBSCAN (eps={eps_value}, min_samples={min_samples_value})\n"
    f"{num_clusters} clusters, {num_outliers} outliers",
    fontsize=16
)

plt.xlabel("Precio/Kilometraje")
plt.ylabel("Año")
plt.legend(title="Cluster", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()

# =========================
# CONCLUSIÓN DINÁMICA
# =========================
if num_clusters > 0:
    conclusion = f"""
## Conclusión DBSCAN

El algoritmo DBSCAN aplicado sobre los datos reducidos con PCA detectó {num_clusters} clusters principales y {num_outliers} outliers.  
Parámetros utilizados: eps={eps_value}, min_samples={min_samples_value}.  
Este formato con paleta Spectral permite distinguir claramente los clusters y resaltar los outliers en rojo.  
Se recomienda seguir ajustando `eps` y `min_samples` para optimizar la separación.
"""
else:
    conclusion = f"""
## Conclusión DBSCAN

El algoritmo DBSCAN no logró detectar clusters válidos y clasificó todos los puntos como ruido (outliers).  
Parámetros utilizados: eps={eps_value}, min_samples={min_samples_value}.  
La visualización con paleta Spectral muestra claramente la dispersión de los datos.  
Se recomienda ajustar los parámetros o considerar otro algoritmo de clustering (como K-Means).
"""

display(Markdown(conclusion))



```

```python
# =========================
# LIBRERÍAS
# =========================
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
from pyspark.sql.types import IntegerType, DoubleType
from pyspark.ml.feature import VectorAssembler, StandardScaler, PCA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import DBSCAN
from IPython.display import Markdown, display

# =========================
# CREAR SESIÓN SPARK
# =========================
spark = (
    SparkSession.builder
    .appName("Clustering_DBSCAN_AutoTec")
    .config("spark.jars.packages", "org.mongodb.spark:mongo-spark-connector_2.12:10.1.1")
    .config("spark.mongodb.read.connection.uri", "mongodb+srv://neiel_cortes:neiel0330@cluster0.eo0kyfv.mongodb.net/AutoTec_db")
    .getOrCreate()
)

# =========================
# LEER DATOS
# =========================
df = spark.read.format("mongodb") \
    .option("database", "proyecto_bigdata") \
    .option("collection", "Contenedor_Autos_Limpio") \
    .load()

print("Registros cargados:", df.count())
df.printSchema()

# =========================
# LIMPIEZA Y CASTEO
# =========================
df_cast = df.withColumn("kilometraje", col("kilometraje").cast(IntegerType())) \
            .withColumn("year", col("year").cast(IntegerType())) \
            .withColumn("precio_num", col("precio").cast(DoubleType())) \
            .dropna(subset=["precio_num", "kilometraje", "year"])

# =========================
# VECTORIZACIÓN Y PCA
# =========================
assembler = VectorAssembler(
    inputCols=["precio_num", "kilometraje", "year"],
    outputCol="features"
)
df_vector = assembler.transform(df_cast)

scaler = StandardScaler(inputCol="features", outputCol="scaledFeatures")
scaler_model = scaler.fit(df_vector)
df_scaled = scaler_model.transform(df_vector)

pca = PCA(k=2, inputCol="scaledFeatures", outputCol="pcaFeatures")
pca_model = pca.fit(df_scaled)
df_pca = pca_model.transform(df_scaled)

# =========================
# EXPORTAR A PANDAS
# =========================
pdf_dbscan = df_pca.select("pcaFeatures", "marca").toPandas()
X = np.array(pdf_dbscan["pcaFeatures"].apply(lambda x: x.toArray()).tolist())

# =========================
# DBSCAN (eps=0.8, min_samples=10)
# =========================
eps_value = 0.8
min_samples_value = 10

dbscan = DBSCAN(eps=eps_value, min_samples=min_samples_value)
clusters = dbscan.fit_predict(X)
pdf_dbscan["cluster_dbscan"] = clusters

# =========================
# VISUALIZACIÓN MEJORADA
# =========================
plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")

# Crear columnas PC1 y PC2
pdf_dbscan[["PC1", "PC2"]] = pd.DataFrame(X, columns=["PC1", "PC2"])

# Paleta más diferenciada
palette = sns.color_palette("Spectral", len(set(clusters)))

# Scatter principal
sns.scatterplot(
    data=pdf_dbscan,
    x="PC1",
    y="PC2",
    hue="cluster_dbscan",
    palette=palette,
    s=80,
    alpha=0.7,
    edgecolor="black"
)

# Resaltar outliers (-1)
outliers = pdf_dbscan[pdf_dbscan["cluster_dbscan"] == -1]
plt.scatter(outliers["PC1"], outliers["PC2"], 
            color="red", s=40, marker="x", label="Ruido (-1)")

# Calcular métricas
num_clusters = len(set(clusters)) - (1 if -1 in clusters else 0)
num_outliers = list(clusters).count(-1)

# Título con parámetros y métricas
plt.title(
    f"Clustering DBSCAN (eps={eps_value}, min_samples={min_samples_value})\n"
    f"{num_clusters} clusters, {num_outliers} outliers",
    fontsize=16
)

# Etiquetas personalizadas de ejes
plt.xlabel("Precio/Kilometraje")
plt.ylabel("Año")

plt.legend(title="Cluster", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()

# =========================
# CONCLUSIÓN DINÁMICA
# =========================
if num_clusters > 0:
    conclusion = f"""
## Conclusión DBSCAN

Con parámetros eps={eps_value} y min_samples={min_samples_value}, el algoritmo DBSCAN detectó {num_clusters} clusters principales y {num_outliers} outliers.  
Esto confirma que DBSCAN puede segmentar vehículos en función de precio, kilometraje y año, aunque la cantidad de outliers depende de la densidad de los datos.  
La visualización mejorada permite distinguir claramente los clusters y resaltar los outliers con símbolos y colores diferenciados.
"""
else:
    conclusion = f"""
## Conclusión DBSCAN

Con parámetros eps={eps_value} y min_samples={min_samples_value}, el algoritmo DBSCAN no logró detectar clusters válidos y clasificó todos los puntos como ruido (outliers).  
Esto puede deberse a la dispersión de los datos o a parámetros poco adecuados.  
Se recomienda seguir ajustando eps y min_samples o considerar otro algoritmo de clustering.
"""

display(Markdown(conclusion))


```
