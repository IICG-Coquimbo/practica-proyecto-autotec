# Evaluación Teórica del Análisis de Clustering

## 1. PCA (Análisis de Componentes Principales)
El PCA es una técnica que permite reducir la cantidad de variables sin perder la esencia de la información. En este proyecto se trabajó con tres variables principales: precio, kilometraje y año de fabricación. Cada vehículo puede pensarse como un punto en un espacio tridimensional, lo que dificulta la visualización y el análisis.  
El PCA transforma estas tres variables en nuevas combinaciones llamadas componentes principales. Los dos primeros componentes (PC1 y PC2) concentran la mayor parte de la variabilidad de los datos, por lo que se decidió trabajar con ellos. De esta manera, se pasó de un espacio de tres dimensiones a un plano de dos dimensiones, lo que facilita la interpretación y la representación gráfica de los clusters.  
La reducción de dimensionalidad permitió simplificar el análisis, eliminar redundancias y visualizar patrones de manera más clara, sin perder información relevante.

## 2. K-Means
El algoritmo K-Means se basa en dividir los datos en grupos homogéneos llamados clusters. Cada cluster tiene un centroide, y los vehículos se asignan al centroide más cercano. El proceso es iterativo: los centroides se recalculan hasta que los grupos quedan bien definidos.  
Este método es eficiente y escalable, lo que lo hace adecuado para trabajar con los 1955 registros del dataset. Sus principales ventajas son la rapidez, la claridad de los resultados y la facilidad de interpretación. Sin embargo, presenta limitaciones: requiere definir el número de clusters antes de ejecutarse y no detecta outliers, ya que todos los datos deben pertenecer a algún grupo.  
En este análisis, K-Means permitió segmentar el mercado en tres grupos principales, lo que aporta una visión clara de cómo se comportan los vehículos según su antigüedad, kilometraje y precio.

## 3. Método del Codo
Para determinar el número óptimo de clusters se utilizó el Método del Codo. Este método se basa en la inercia, que mide la suma de las distancias entre los puntos y sus centroides.  
Al graficar la inercia contra el número de clusters, se observa que al inicio la inercia disminuye rápidamente, pero llega un punto en que la mejora se vuelve marginal. Ese punto se denomina “codo” y representa el número óptimo de clusters.  
En este caso, el codo se identificó en **k=3**, lo que justificó la elección de tres clusters como la mejor opción para representar la estructura de los datos.

## 4. Silhouette Score
El Silhouette Score es una métrica que evalúa la calidad de los clusters. Considera tanto la cohesión interna (qué tan cerca están los puntos dentro de un mismo cluster) como la separación externa (qué tan alejados están de otros clusters).  
Su rango de interpretación va de -1 a 1: valores cercanos a 1 indican clusters bien definidos, valores cercanos a 0 muestran solapamiento y valores negativos reflejan asignaciones incorrectas.  
En este análisis se obtuvo un valor de **0.412**, lo que indica que los clusters son aceptables. Existe cierta mezcla en los límites, pero la separación general es válida y suficiente para justificar la segmentación.

## 5. DBSCAN
El algoritmo DBSCAN agrupa datos en función de su densidad. Si un punto tiene suficientes vecinos cercanos, se considera parte de un cluster; si no, se clasifica como ruido.  
Su principal ventaja es que no requiere definir el número de clusters previamente y puede detectar outliers de manera automática. Además, permite formar clusters de formas irregulares, lo que lo hace más flexible que K-Means.  
En este proyecto, con parámetros `eps=0.7` y `min_samples=3`, DBSCAN identificó **2 clusters principales y 4 outliers**. Estos outliers corresponden a vehículos con características extremas, como precios o kilometrajes fuera de lo común, lo que aporta valor adicional al análisis.

## 6. Resultados obtenidos
El análisis permitió identificar tres segmentos principales en el mercado de vehículos usados:  
- **Cluster de mayor depreciación:** autos antiguos y con alto kilometraje, caracterizados por precios bajos.  
- **Cluster intermedio:** autos con características equilibradas, con precios medios y un balance entre antigüedad y kilometraje.  
- **Cluster premium:** autos recientes y con bajo kilometraje, que representan el segmento de mayor valor comercial.  
El complemento de DBSCAN permitió confirmar la existencia de vehículos atípicos, resaltando la importancia de considerar outliers en futuros modelos predictivos.

---

## Conclusión General
El estudio confirma que **kilometraje y antigüedad son factores determinantes en la depreciación de vehículos usados**.  
El uso de **K-Means** permitió una segmentación clara y estable en tres grupos, mientras que **DBSCAN** complementó el análisis al identificar autos fuera de rango.  
En conjunto, ambas técnicas ofrecen una visión integral del mercado automotriz: por un lado, la estructura principal de los segmentos y, por otro, la detección de casos especiales que podrían influir en estrategias de compra, venta y en futuros modelos de predicción de precios.
