## Explicación del código DBSCAN

### 1. Importación de librerías
Se cargan librerías necesarias:
- **Spark**: para manejar datos grandes.
- **Pandas/Numpy**: para trabajar con tablas y arrays.
- **Matplotlib/Seaborn**: para graficar.
- **DBSCAN**: el algoritmo de clustering que se va a aplicar.

### 2. Sesión Spark y lectura de datos
Se crea una sesión de Spark y se leen los datos desde MongoDB.  
El dataset final tiene **1955 registros limpios** con las variables:
- `precio`
- `kilometraje`
- `year`

### 3. Limpieza y casteo
Se convierten las columnas a valores numéricos (`precio_num`, `kilometraje`, `year`) y se eliminan los nulos.  
Esto asegura que DBSCAN trabaje con datos consistentes.

### 4. Vectorización y escalado
- Se juntan las variables en un vector (`features`).  
- Se escalan los datos para que todas las variables tengan el mismo peso.  
- Se aplica **PCA** (reducción de dimensiones) para graficar en 2D.

### 5. Aplicación de DBSCAN
- Parámetros óptimos: `eps=0.7`, `min_samples=3`.  
- DBSCAN detecta **2 clusters principales** y **4 outliers**.  
- Los outliers se marcan como `-1` (ruido).

### 6. Visualización (el ultimo es el con datos optimos)
- Los clusters se grafican con colores distintos.  
- Los outliers aparecen en rojo con una “x”.  
- El título muestra los parámetros y resultados.

### 7. Conclusión automática
El código imprime un resumen:
- Clusters detectados: 2  
- Outliers: 4  
- Silhouette Score: 0.711  

Esto confirma que la configuración elegida separa bien los grupos y el ruido.

## Explicación del código LOOP (dbscan_pruebas)

Este bloque sirve para **probar varias combinaciones de parámetros** y ver cómo cambia el resultado.

### 1. Subplots 2x2
Se crean 4 gráficos en una sola figura:
- Arriba izquierda: número de clusters según `eps` y `min_samples`.  
- Arriba derecha: número de outliers según `eps` y `min_samples`.  
- Abajo izquierda: heatmap del **Silhouette Score** (calidad del clustering).  
- Abajo derecha: heatmap de cantidad de outliers.

### 2. Interpretación de los gráficos
- El heatmap del Silhouette Score muestra que la mejor calidad se logra con `eps=0.7` y `min_samples=3`.  
- El heatmap de outliers muestra que esa configuración detecta pocos outliers (4).  
- Los gráficos de líneas permiten ver cómo cambian clusters y outliers cuando se ajustan los parámetros.

### 3. Conclusión automática del loop
El código selecciona la mejor configuración y la imprime:
- eps=0.7, min_samples=3  
- Clusters: 2  
- Outliers: 4  
- Silhouette Score: 0.711  

Esto confirma que esa combinación es la más adecuada.
