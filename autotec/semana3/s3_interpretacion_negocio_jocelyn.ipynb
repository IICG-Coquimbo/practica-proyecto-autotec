# Interpretación de Resultados EDA - AutoTec

## Objetivo del análisis

El objetivo del análisis exploratorio de datos fue comprender el comportamiento de las principales variables asociadas al mercado de vehículos usados presentes en la base de datos AutoTec. Para ello se analizaron características como precio, año de fabricación, kilometraje, marca y modelo, buscando identificar patrones relevantes para la toma de decisiones basada en datos.

## Hallazgos principales

Los resultados obtenidos muestran que existe una relación entre el precio de los vehículos y variables como el año de fabricación y el kilometraje. En términos generales, los vehículos más recientes presentan precios más elevados, mientras que aquellos con mayores niveles de uso tienden a disminuir su valor comercial.

Asimismo, el análisis permitió identificar diferencias entre marcas y modelos, evidenciando que ciertos segmentos mantienen una valorización superior dentro del mercado automotriz usado. Estas diferencias reflejan el comportamiento natural de la oferta y demanda, así como la percepción de valor asociada a determinadas marcas.

Por otra parte, la detección de valores atípicos permitió identificar vehículos cuyos precios se encuentran considerablemente alejados del comportamiento promedio del conjunto de datos. Estos casos pueden corresponder a automóviles de alta gama, modelos exclusivos o vehículos con características especiales.

## Impacto para AutoTec

Desde una perspectiva de negocio, los resultados obtenidos permiten comprender mejor los factores que influyen en el precio de los vehículos usados. Esta información puede ser utilizada para apoyar procesos de valorización, segmentación de inventario y análisis comparativo entre distintos tipos de automóviles.

Además, el conocimiento generado a partir del análisis exploratorio constituye una base sólida para futuras etapas de minería de datos, permitiendo implementar técnicas de clustering y modelos predictivos orientados a la estimación de precios y clasificación de vehículos.

## Conclusión

El análisis exploratorio permitió identificar patrones relevantes dentro del mercado automotriz usado y validar la importancia de variables como precio, kilometraje, año, marca y modelo. Los resultados obtenidos proporcionan información útil para la toma de decisiones y respaldan el desarrollo de soluciones analíticas más avanzadas en las siguientes etapas del proyecto AutoTec.
