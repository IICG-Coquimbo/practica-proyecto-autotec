PROPUESTA DE VALOR AUTOTEC:    

El presente proyecto entrega a las organizaciones del sector automotriz una solución basada en datos para resolver uno de sus problemas más costosos: la tasación imprecisa de vehículos usados. A través del análisis multivariado del mercado, se identificó que el precio de un vehículo no depende de una sola variable, sino de la interacción entre varias como lo son  el kilometraje, la antigüedad, la marca, el modelo y el tipo de combustible. Ignorar esta combinación —como ocurre en los métodos manuales actuales— genera sistemáticamente subvaloraciones y sobrevaloracioes que afectan directamente la rentabilidad y competitividad del negocio
La propuesta de valor concreta es reemplazar el criterio subjetivo por un modelo de decisión estructurado en tres niveles: estratégico, táctico y operativo. A nivel estratégico, permite a la gerencia entender cómo se comporta el mercado en su conjunto y qué marcas lideran la oferta. A nivel táctico, entrega a las jefaturas herramientas para fijar precios competitivos según combustible, nivel de uso y posicionamiento de marca. A nivel operativo, alerta en tiempo real a los tasadores sobre qué vehículos presentan desviaciones críticas respecto al precio de mercado esperado, priorizando la revisión manual donde el riesgo económico es mayor.
El resultado es una automotora que compra mejor, vende con mayor precisión y reduce su exposición al riesgo financiero derivado de tasaciones incorrectas, todo ello sustentado en evidencia del mercado real y no en suposicio.s.

Interpretación
El gráfico revela que el tipo de combustible tiene un impacto directo y diferenciado sobre el precio según el nivel de uso del vehículo. 
Los vehículos híbridos mantienen los precios promedio más altos en todos los rangos de kilometraje, lo que refleja una mayor retención
de valor asociada tanto a su tecnología como a su demanda en el mercado. Los vehículos diésel presentan un comportamiento intermedio: 
compiten bien en el tramo de uso bajo, pero su valor cae de forma más pronunciada en kilometrajes altos. Los vehículos a bencina, siendo los 
más abundantes en el mercado, muestran la mayor sensibilidad al nivel de uso: su precio decrece de forma sostenida a medida que el
kilometraje aumenta. En todos los tipos de combustible, el paso del rango Bajo al Alto implica una caída significativa de precio, 
confirmando que el nivel de uso es un factor de depreciación transversal, independiente del tipo de motor.

Propuesta Estratégica
Priorizar la adquisición de vehículos híbridos de uso bajo y medio: son los que mejor retienen valor y permiten márgenes de reventa 
más competitivos para la automotora.
Establecer políticas de precio diferenciadas por tipo de combustible y rango de kilometraje: un vehículo diésel de uso alto no debe 
tasarse con el mismo criterio que uno de uso bajo, ya que el mercado los valora de forma distinta.
Utilizar esta combinación de variables (combustible y nivel de uso) como criterio base en el proceso de tasación táctica, reduciendo la 
dependencia de criterios subjetivos al momento de fijar precios de compra y venta.



GRÁFICO 2: Posicionamiento de precios por marca 

Descripción
Este gráfico (boxplot) muestra la distribución de precios de los vehículos según cada marca. El eje horizontal presenta las marcas y el eje vertical los precios en pesos chilenos. Para cada marca se visualiza la mediana, el rango intercuartílico (que concentra el 50% central de los precios) y los valores atípicos representados como puntos fuera de los bigotes. Las marcas se ordenan según su precio mediano, lo que facilita comparar su posicionamiento relativo en el mercado
.
Interpretación
El gráfico permite identificar dos grandes grupos de marcas. Por un lado, marcas con una banda de precios amplia y dispersa —como Ford, Chevrolet, Toyota o Volkswagen— cuya variación no se explica solo por el kilometraje o el año, sino principalmente por el modelo específico: un mismo fabricante puede ofrecer desde un compacto de entrada hasta un SUV/camioneta de alta gama, lo que genera rangos de precio muy distintos dentro de una misma marca. Por otro lado, marcas como Chery, MG o Peugeot muestran bandas más estrechas y concentradas, lo que indica una oferta más homogénea y una menor variabilidad en su precio de mercado. Esta dispersión diferenciada representa un riesgo directo de tasación: usar el precio promedio de la marca como referencia en marcas con alta dispersión puede derivar en subvaloración o sobrevaloración si no se considera el model
o.
Propuesta Estratégica
Mantener o reforzar el posicionamiento premium en marcas con precio promedio alto y rango controlado: su consistencia de precios facilita la tasación precisa y reduce el riesgo de error.
Revisar el pricing en marcas con banda muy amplia incorporando el modelo como variable obligatoria en la tasación: no basta con identificar la marca; el modelo determina en qué segmento del rango de precios opera ese vehículo específico.
Ajustar el mix de inventario priorizando marcas con mejor relación entre valor promedio y consistencia de precio: ya que permiten procesos comerciales más predecibles y menor exposición a errores de valoración.


GRÁFICO 3: Retención de valor por marca 

Descripción
Este gráfico de barras horizontales muestra un índice de retención de valor calculado para cada marca, donde el valor 100 representa el punto de equilibrio óptimo de mercado. Un índice inferior a 100 indica que la marca se deprecia más rápido que el promedio del mercado, mientras que un valor superior a 100 señala que la marca mantiene —o incluso supera— el valor esperado según su uso y antigüedad. El eje horizontal presenta el valor del índice y el eje vertical las marcas ordenadas de mayor a menor retención
.
Interpretación
El índice revela diferencias importantes en la velocidad de depreciación entre marcas. Las marcas con índice superior a 100 logran mantener precios por encima de lo que su nivel de uso y antigüedad justificaría en promedio, lo que las convierte en activos más rentables para la automotora. En contraste, las marcas con índice inferior a 100 pierden valor más rápido de lo esperado: tasarlas con criterios generales sin considerar su comportamiento específico de depreciación genera un riesgo real de sobre valorarlas al momento de la compra. Este gráfico es especialmente útil para decisiones de compra de inventario, ya que permite anticipar qué marcas ofrecerán mejor retorno al momento de la reven
ta.
Propuesta Estratégica
Priorizar la compra de vehículos de marcas con índice superior a 100: su mayor retención de valor protege el margen de la automotora y reduce el riesgo de pérdida al momento de la reventa.
Aplicar descuentos de compra más agresivos en marcas con índice bajo: si la marca se deprecia rápido, el precio de adquisición debe reflejar esa realidad para preservar el margen comercial.
Incorporar el índice de retención como criterio de evaluación periódica del portafolio: marcas que caen bajo el umbral de 100 de forma sostenida deben revisarse estratégicamente para decidir si se mantienen, se renegocian las condiciones de compra o se priorizan otras alternativas con mejor comportamiento de mercado.



CONCLUSIÓN AUTOTEC

El análisis desarrollado a lo largo de este proyecto confirmó que la valoración de vehículos usados es un problema multivariable que no puede resolverse con criterios únicos ni métodos manuales. Los datos demostraron que el precio de mercado de un vehículo es el resultado de la interacción entre varias variables  y que ignorar cualquiera de estas abre una brecha entre el precio tasado y el precio real que el mercado está dispuesto a pagar.
El enfoque por niveles —estratégico, táctico y operativo— permitió traducir esa complejidad en información accionable para cada capa de la organización, desde la gerencia hasta el tasador en terreno. Esto es precisamente lo que diferencia este proyecto de un análisis tradicional: no solo describe el mercado, sino que entrega decisiones concretas adaptadas a quien las necesita tomar.
Como hallazgo central, se estableció que el kilometraje es el principal motor de depreciación, pero su impacto varía significativamente según la marca y el tipo de combustible, lo que invalida cualquier modelo de tasación que opere con promedios generales. Las marcas con alta dispersión de precio y los vehículos en zona de alta depreciación representan los mayores focos de riesgo financiero para la organización, y son precisamente donde una herramienta de decisión basada en datos genera mayor valor.
En definitiva, este proyecto le entrega las herramientas necesarias para dejar de fijar precios basándose en experiencia o criterios generales, y pasar a tomar decisiones de tasación fundamentadas en el comportamiento real del mercado.
