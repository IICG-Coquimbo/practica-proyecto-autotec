```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, regexp_replace, when, lit
from pyspark.sql.types import FloatType, IntegerType

spark = SparkSession.builder \
    .appName("EDA_AutoTec_Dani_Kilometraje") \
    .config("spark.mongodb.read.connection.uri", "mongodb+srv://neiel_cortes:neiel0330@cluster0.eo0kyfv.mongodb.net/AutoTec_db") \
    .config("spark.jars.packages", "org.mongodb.spark:mongo-spark-connector_2.12:10.1.1") \
    .getOrCreate()

df_raw = spark.read.format("mongodb") \
    .option("database", "proyecto_bigdata") \
    .option("collection", "lista_autos") \
    .load()

print(df_raw.count())
```

```python
df_cleaned = df_raw.filter(
    (col("kilometraje").isNotNull()) &
    (col("kilometraje") != "") &
    (col("kilometraje") != "Sin kilometraje") &
    (col("kilometraje") != "Sin información")
)

df_cleaned.select("marca", "modelo", "kilometraje").show(10, truncate=False)

print(f"Total de registros después de limpiar kilometraje: {df_cleaned.count()}")
```

```python
df_normalized = df_cleaned.withColumn(
    "kilometraje_clean",
    regexp_replace(col("kilometraje"), r"[^\d]", "")
)
```

```python
df_kilometraje = df_normalized.withColumn(
    "kilometraje_num",
    col("kilometraje_clean").cast(FloatType())
)

df_kilometraje.select("marca", "modelo", "kilometraje", "kilometraje_num").show(10, truncate=False)

df_kilometraje.printSchema()
```

```python
df_kilometraje = df_kilometraje.withColumn(
    "rango_kilometraje",
    when(col("kilometraje_num") < 50000, "Bajo")
    .when((col("kilometraje_num") >= 50000) & (col("kilometraje_num") < 120000), "Medio")
    .otherwise("Alto")
)

df_kilometraje.select(
    "marca",
    "modelo",
    "kilometraje_num",
    "rango_kilometraje"
).show(10)
```

```python
df_kilometraje.filter(
    (col("kilometraje_num") < 1000) |
    (col("kilometraje_num") > 300000)
).select(
    "marca",
    "modelo",
    "year",
    "kilometraje_num",
    "precio"
).show(20, truncate=False)
```

```python
df_kilometraje.select("kilometraje_num").describe().show()
```

```python
from pyspark.sql.functions import avg, count, round

df_kilometraje.groupBy("marca") \
    .agg(
        count("*").alias("N_Autos"),
        round(avg("kilometraje_num"), 2).alias("Avg_Kilometraje")
    ) \
    .orderBy(col("Avg_Kilometraje").desc()) \
    .show(20, truncate=False)
```

```python
df_kilometraje_clustering = df_kilometraje.select(
    "marca",
    "modelo",
    "year",
    "precio",
    "kilometraje",
    "kilometraje_num",
    "rango_kilometraje"
)

df_kilometraje_clustering.show(10, truncate=False)
```

```python
df_kilometraje_clustering.write \
    .format("mongodb") \
    .mode("overwrite") \
    .option(
        "spark.mongodb.write.connection.uri",
        "mongodb+srv://neiel_cortes:neiel0330@cluster0.eo0kyfv.mongodb.net/AutoTec_db"
    ) \
    .option("database", "proyecto_bigdata") \
    .option("collection", "Kilometraje_Dani_Limpio") \
    .save()
```
