# Punto 4: Segmentación mediante Clustering

El objetivo de esta etapa es identificar segmentos de vehículos con características similares mediante técnicas de aprendizaje no supervisado.

La segmentación permite comprender mejor el comportamiento del mercado de vehículos usados y analizar patrones de depreciación considerando variables relevantes como precio, kilometraje y año de fabricación.

Los resultados obtenidos permitirán identificar grupos de vehículos con distintos niveles de valor comercial, antigüedad y desgaste, aportando información relevante para futuras etapas de análisis predictivo.

```python
df_clustering = df.select(
    col("precio").cast("double"),
    col("kilometraje").cast("double"),
    col("year").cast("int").alias("anio")
)
```

```python
df_clustering = df_clustering.dropna()

df_clustering = df_clustering.filter(col("kilometraje") < 800000)
df_clustering = df_clustering.filter(col("precio") > 0)
df_clustering = df_clustering.filter(col("anio") >= 1990)

print("Registros después de limpieza:", df_clustering.count())

df_clustering.show(5)
```

```python
from pyspark.ml.feature import VectorAssembler, StandardScaler, PCA

assembler = VectorAssembler(
    inputCols=["precio", "kilometraje", "anio"],
    outputCol="features"
)

df_vector = assembler.transform(df_clustering)

scaler = StandardScaler(
    inputCol="features",
    outputCol="scaledFeatures"
)

scaler_model = scaler.fit(df_vector)

df_scaled = scaler_model.transform(df_vector)

pca = PCA(
    k=2,
    inputCol="scaledFeatures",
    outputCol="pcaFeatures"
)

pca_model = pca.fit(df_scaled)

df_pca = pca_model.transform(df_scaled)
```

```python
pdf_pca = df_pca.select("pcaFeatures").toPandas()

pdf_pca[["PC1", "PC2"]] = pd.DataFrame(
    pdf_pca["pcaFeatures"].apply(lambda x: x.toArray()).tolist()
)

plt.figure(figsize=(10, 7))

sns.scatterplot(
    data=pdf_pca,
    x="PC1",
    y="PC2",
    color="teal",
    alpha=0.6,
    s=60
)

plt.title("Distribución Espacial de Vehículos mediante PCA")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")

plt.show()
```

```python
from pyspark.ml.clustering import KMeans
```

```python
cost = []

for k in range(2, 11):

    kmeans = KMeans(
        featuresCol="scaledFeatures",
        k=k,
        seed=42
    )

    model = kmeans.fit(df_scaled)

    cost.append(model.summary.trainingCost)
    
```

```python
cost
```

## Selección y Justificación del Algoritmo

Durante el desarrollo del proyecto se evaluaron distintas técnicas de clustering, principalmente K-Means y DBSCAN.

Tras revisar los resultados obtenidos y comparar ambas alternativas, se decidió utilizar K-Means como algoritmo principal debido a que permite generar grupos claramente diferenciados e interpretables desde una perspectiva comercial.

Las variables utilizadas para la segmentación fueron precio, kilometraje y año de fabricación, ya que representan factores fundamentales en el análisis de la depreciación de vehículos usados.

DBSCAN fue utilizado como técnica complementaria para identificar posibles anomalías y validar la estructura general de los datos.

```python
plt.figure(figsize=(8,5))

plt.plot(
    range(2,11),
    cost,
    marker='o'
)

plt.xlabel("Número de Clusters (K)")
plt.ylabel("Costo / Inercia")
plt.title("Método del Codo - AutoTec")

plt.show()
```

## Validación Técnica: Método del Codo

Para determinar la cantidad óptima de clústeres se aplicó el Método del Codo.

La inercia disminuye significativamente entre k=2 y k=3. A partir de este punto la reducción comienza a ser menos pronunciada, indicando que agregar más grupos aporta una mejora marginal en la segmentación.

Por esta razón se seleccionó k=3 como número óptimo de clústeres para representar los distintos segmentos presentes en el mercado de vehículos usados.

```python
k_optimo = 3
```

```python
kmeans_final = KMeans(
    featuresCol="scaledFeatures",
    k=k_optimo,
    seed=42
)

model_final = kmeans_final.fit(df_scaled)

df_clusters = model_final.transform(df_scaled)
```

```python
df_clusters.select(
    "precio",
    "kilometraje",
    "anio",
    "prediction"
).show(20)
```

```python
df_clusters.groupBy("prediction").count().show()
```

```python
df_viz = model_final.transform(df_pca)

pdf_visual = df_viz.select(
    "pcaFeatures",
    "prediction"
).toPandas()

pdf_visual[["PC1","PC2"]] = pd.DataFrame(
    pdf_visual["pcaFeatures"].apply(
        lambda x: x.toArray()
    ).tolist()
)
```

```python
plt.figure(figsize=(12,8))

sns.scatterplot(
    data=pdf_visual,
    x="PC1",
    y="PC2",
    hue="prediction",
    palette="viridis",
    s=80,
    alpha=0.7
)

plt.title("Clusters K-Means en Espacio PCA")

plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")

plt.show()
```

## Visualización de Clústeres mediante PCA

Para facilitar la interpretación de los resultados obtenidos por K-Means se utilizó la técnica de reducción de dimensionalidad PCA (Análisis de Componentes Principales).

La visualización muestra la distribución de los vehículos en tres segmentos identificados por el algoritmo. Cada color representa un grupo de vehículos con características similares considerando variables relevantes para el análisis de depreciación, como precio, kilometraje y año.

La separación observada entre los grupos indica que existen distintos perfiles de vehículos dentro del mercado analizado, permitiendo identificar segmentos con diferentes niveles de valor comercial y depreciación.


```python
df_clusters.groupBy("prediction").avg(
    "precio",
    "kilometraje",
    "anio"
).show()
```

## Perfilamiento de Clústeres

### Clúster 1: Vehículos con mayor depreciación

Este grupo presenta el menor precio promedio ($11,25 millones), el mayor kilometraje promedio (125.435 km) y una antigüedad superior respecto a los demás segmentos (año promedio 2017).

Estos vehículos representan el segmento con mayor depreciación acumulada, ya que el elevado kilometraje y la antigüedad reducen significativamente su valor comercial.

### Clúster 0: Vehículos de valor intermedio

Este segmento posee un precio promedio de $14,31 millones, un kilometraje promedio de 44.694 km y un año promedio 2022.

Corresponde a vehículos relativamente recientes que mantienen una depreciación moderada y un valor comercial intermedio dentro del mercado.

### Clúster 2: Vehículos de mayor valor comercial

Este grupo presenta el precio promedio más alto ($37,58 millones), el menor kilometraje promedio (42.649 km) y el año promedio más reciente (2023).

Estos vehículos representan el segmento premium o de mayor valor dentro del mercado analizado, mostrando una depreciación significativamente menor en comparación con los otros grupos.


## Interpretación de Negocio

Los resultados obtenidos confirman que el kilometraje y el año de fabricación son variables fundamentales para explicar la depreciación de los vehículos usados.

Se observa una relación directa entre antigüedad, kilometraje y pérdida de valor comercial. Los vehículos más antiguos y con mayor kilometraje presentan los menores precios promedio, mientras que los vehículos más recientes conservan una mayor proporción de su valor original.

La segmentación obtenida permite identificar distintos perfiles de vehículos dentro del mercado, facilitando futuras estrategias de valorización, compra, venta y predicción de precios.


```python
from sklearn.cluster import DBSCAN
```

```python
pdf_dbscan = df_pca.select(
    "pcaFeatures"
).toPandas()

X = np.array(
    pdf_dbscan["pcaFeatures"].apply(
        lambda x: x.toArray()
    ).tolist()
)
```

```python
dbscan = DBSCAN(
    eps=0.5,
    min_samples=5
)

clusters_dbscan = dbscan.fit_predict(X)

pdf_dbscan["cluster_dbscan"] = clusters_dbscan
```

```python
pdf_dbscan["cluster_dbscan"].value_counts()
```

```python
pdf_dbscan[["PC1", "PC2"]] = pd.DataFrame(
    X,
    columns=["PC1", "PC2"]
)

plt.figure(figsize=(12, 8))

sns.scatterplot(
    data=pdf_dbscan,
    x="PC1",
    y="PC2",
    hue="cluster_dbscan",
    palette="Set1",
    s=70,
    alpha=0.6
)

plt.title("Clustering DBSCAN - AutoTec")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.legend(title="Cluster DBSCAN (-1 = Ruido)")

plt.show()
```

## Interpretación de DBSCAN

DBSCAN fue utilizado como técnica complementaria para detectar anomalías y validar la estructura de los datos.

Los resultados muestran un grupo principal compuesto por la mayoría de los registros y un pequeño conjunto de observaciones catalogadas como ruido o anomalías.

Estas anomalías pueden corresponder a vehículos con precios inusualmente altos o bajos, kilometrajes extremos o características poco frecuentes respecto al comportamiento general del mercado.

La baja cantidad de registros atípicos detectados sugiere una buena calidad de los datos utilizados en el análisis.

```python
from pyspark.ml.evaluation import ClusteringEvaluator

evaluator = ClusteringEvaluator(
    predictionCol="prediction",
    featuresCol="scaledFeatures",
    metricName="silhouette"
)

silhouette = evaluator.evaluate(df_clusters)

print("Silhouette Score:", silhouette)
```

## Validación Técnica: Silhouette Score

Para complementar el Método del Codo se calculó el índice de Silueta del modelo K-Means final.

El resultado obtenido fue:

**Silhouette Score = 0.412**

Este valor indica que existe una separación moderada entre los clústeres. Si bien no representa una separación perfecta, sí permite identificar segmentos diferenciados de vehículos usados según precio, kilometraje y año.

Por esta razón, se mantiene K-Means con k=3 como modelo final, ya que entrega grupos interpretables desde una perspectiva de negocio: vehículos con mayor depreciación, vehículos de valor intermedio y vehículos de mayor valor comercial.
